# **Data Visualization**

## Why Build Visuals?

Well data visualization is a way to show a complex data in a form that is graphical and easy to understand. This can be especially useful when one is trying to explore the data and getting acquainted with it. 

- For exploratory data analysis
- Communicate data clearly
- Share unbiased representation of data
- Use them to support recommendations to different stakeholders.

### <font color='green'>Best Practices</font>

When creating a visual, always remember :
- LESS IS MORE EFFECTIVE
- LESS IS MORE ATTRACTIVE
- LESS IS MORE IMPACTIVE

<font color='dark orange'>Always keep this thing in your mind that whatever the message you want to give must be very clear with your plot, don't distract from it while making the plot more appealing.</font>

Let's take a look at an example. So here is a pie chart of `what looks like people's preferences when it comes to different types of pig meat`. The charts message is almost half of the people surveyed preferred bacon over the other types of pig meat. But I'm sure that almost all of you agree that there is a lot going on in this pie chart and we're not even sure it features such as the blue background or 3d orientation are meant to convey anything. In fact, `these additional unnecessary features distract from the main message and can be confusing to the audience.`

**So let's apply Darkhorse Analytics approach to transform this into a visual that's more effective, attractive, and impactive**

### <font color='BLUE'> SO CAN ANYONE TELL ME WHAT IS THE MAIN MESSAGE HERE?</font>

*the message here is that people are most likely to choose bacon over other types of pig meat, so let's get rid of everything that can be distracting from this core message.*

- The first thing is let's get rid of the `blue background` and the gray background. 
- Let's also get rid of `borders` as they do not convey any extra information. 
- Also let's drop the `redundant legend`(the wording on a map or diagram explaining the symbols used.) since the pie chart is already color coded. 
- `3D` isn't adding any extra information so let's say bye to it. 
- `Text bolding` is also unnecessary, and let's get rid of the different `colors` and the `wedges`.

## Plot Function

```
# importing scripting interface as plt
%matplotlib inline
import matplotlib.pyplot as plt
```

```
# plotting circular mark at the position (5,5)
plt.plot(5,5,'o')
plt.show()
```

Plotting interactively within an IPython notebook can be done with the `%matplotlib` command, and works in a similar way to the IPython shell. In the IPython notebook, you also have the option of embedding graphics directly in the notebook, with two possible options:

- `%matplotlib notebook` will lead to interactive plots embedded within the notebook
- `%matplotlib inline` will lead to static images of your plot embedded in the notebook

**Magic Function** -<font color='purple'> A magic function starts with % Matplotlib, and to enforce plots to be rendered within the browser, you pass in inline as the backend.</font>

Finally, another thing that is great about Matplotlib is that **`pandas also has a built-in implementation of it`**. Therefore, plotting in pandas is as simple as calling the plot function on a given pandas series or dataframe.

Let's see how to do that :-


```
list1 = [8880, 8670, 8147, 7338, 5704, 5846, 5231, 4568, 4213, 3210, 2145, 2987, 1258, 1369, 1478, 1032, 1009]
list2 = [8280, 2382, 2564, 5893, 5675, 6542, 6985, 5231, 4589, 4523, 4102, 3269, 3657, 2653, 4587, 5695, 2546]
list3 = [1980, 1981, 1982, 1983, 1984, 1985, 1986, 1987, 1988, 1989, 1990, 1991, 1992, 1993, 1994, 1995, 1996]
```

```
#creating a dataframe from a above THREE lists
d = {'year':list3,'China':list2,'India':list1}
```

```
import pandas as pd
india_china_df = pd.DataFrame(d)

india_china_df = india_china_df.set_index('year')
india_china_df.index.name = None
india_china_df
```

```
# line plot 
india_china_df.plot(kind = 'line')
```

```
 # histogram
 india_china_df['India'].plot(kind = 'hist')
```
