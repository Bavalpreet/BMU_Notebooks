##Http basics


The "Hyper text transfer protocol" AKA HTTP is the world wide protocol for how we communicate and send/receive information on the web.

It works on a "Request/Response" protocol, where a client makes a request to a web server which then returns a response.

##What are http methods?

Http protocol is the foundation of data communication in world wide web. Different methods of data retrieval from specified URL are defined in this protocol.


1.   **GET**= Used to fetch the specified resource. Most common method and default method in FLASK.

2.   **POST** = Used to create new data at the specified resource.
3.**PUT** = Used to create new data or replace existing data at the specified resource 
4.   **DELETE** =  Used to delele existing data at the specified resource




### HTTP status codes

HTTP status codes are made up of 3 digits that fall into 5 categories, with each category representing a certain class of code.

The first digit is the category and the 5 categories correspond to the following class:

1xx - Informational

2xx - Success

3xx - Redirection

4xx - Client errors

5xx - Server errors

# Now Let's move to <font color='red'>sessions</font> very Important concept





**Session** is the time interval when a client logs into a server and logs out of it. The data, which is needed to be held across this session, is stored in the client browser.

A session with each client is assigned a **Session ID**. The Session data is stored on top of cookies and the server signs them **cryptographically**. For this encryption, a Flask application needs a defined **SECRET_KEY**.

Session object is also a dictionary object containing key-value pairs of session variables and associated values.
